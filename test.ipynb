```python
arr = [11, 6, 13, 9, 15, 17, 18, 19]
n = len(arr)
for i in range(n-1):
    flag = False 
    for j in range(n-i-1):
        if arr[j] > arr[j+1]:
            arr[j], arr[j+1] = arr[j+1], arr[j]
            flag = True
        if not flag:
            break
print(arr)
```

selection sort 
mỗi 1 loop ta đi tìm phần tử nhỏ nhất và để lên đầu
vậy loop đầu ta sẽ chạy hết toàn mảng, vậy ta sẽ có bao nhiều lần tìm min? ta sẽ có n-1 tìm min?

```python
arr = [11, 6, 13, 9, 15, 17, 18, 19]
n = len(arr)
for i in range(n-1):
    flag = False
    minval = i
    for j in range(i+1, n):
        if arr[j] < arr[minval]:
            minval = j
            flag = True
    arr[i], arr[minval] = arr[minval], arr[i]
    if not flag:
        break
    print(arr)

print(arr)
```

```python
array = [11, 6, 13, 9, 15, 17, 18, 19]
def find_min(arr):
    minv = 0
    for i in range(1, len(arr)):
        if arr[i] < arr[minv]:
            minv = i
        else:
            continue
    return arr[minv]
print(find_min(array))
```
